```python
# Import relevant packages

import numpy as np 
from matplotlib import pyplot as pl 
from spectres import spectres
```

```python
# Read the band wavelengths for each instrument

AST_WAVELEN = open('S07ASTER_Wavelengths_ASTER_(9_bands)_microns.txt')
astfusedwave = AST_WAVELEN.readlines()[4:10]
for i in range(0, len(astfusedwave)): 
    astfusedwave[i] = float(astfusedwave[i])   
    
SENT_WAVELEN = open('S07SNTL2_Wavelengths_Sentinel2_(13_bands)_microns.txt')
sentfusedwave = SENT_WAVELEN.readlines()[2:10]
for i in range(0, len(sentfusedwave)): 
    sentfusedwave[i] = float(sentfusedwave[i])
    
    
AST_TIR_WAVELEN = open('splib07a_Wavelengths_NIC4_Nicolet_1.12-216microns.txt')
astfusedtirwave = AST_TIR_WAVELEN.readlines()[922:]
for i in range (0, len(astfusedtirwave)):
    astfusedtirwave[i] = float(astfusedtirwave[i])
    
BEC_WAVE = open('splib07a_Wavelengths_BECK_Beckman_0.2-3.0_microns.txt')
becwaves = BEC_WAVE.readlines()[1:]
for i in range (0, len(becwaves)):
    becwaves[i] = float(becwaves[i])
becwaves = np.array(becwaves)
```

```python
# Fill this whole kernal out for each new mineral to give VNIR - SWIR and TIR reflectance

# Read resampled ASTER and Sentinel-2 reflectance and NIC NICOLET 1.12 - 216µm raw reflectance values

AST_REF = open('S07ASTER_Chlorite_HS179.1B_ASDFRb_AREF.txt')

astfusedrefchl = AST_REF.readlines()[4:10]
for i in range (0, len(astfusedrefchl)):
    astfusedrefchl[i] = float(astfusedrefchl[i])
    
SENT_REF = open('S07SNTL2_Chlorite_HS179.1B_ASDFRb_AREF.txt')

sentfusedrefchl = SENT_REF.readlines()[2:10]
for i in range(0, len(sentfusedrefchl)):
    sentfusedrefchl[i] = float(sentfusedrefchl[i])
    
AST_TIR_REF = open('splib07a_Chlorite_HS179.3B_NIC4bbu_RREF.txt')

astfusedtirrefchl = AST_TIR_REF.readlines()[922:]
for i in range (0, len(astfusedtirrefchl)):
    astfusedtirrefchl[i] = float(astfusedtirrefchl[i])

#-----------------------------------------------------------------------------------------     

# Order the VNIR - SWIR bands by increasing corresponding wavelength

fusedwave = astfusedwave + sentfusedwave
fusedrefchl = astfusedrefchl + sentfusedrefchl

zipped_lists = zip(fusedwave, fusedrefchl)
sorted_pairs = sorted(zipped_lists)

tuples = zip(*sorted_pairs)
fusedwavelength, fusedrefchl = [ list(tuple) for tuple in  tuples]

#----------------------------------------------------------------------------------------- 

# Resample the NIC NICOLET 1.12 - 216µm reflectance data to ASTER TIR bands

TIR = [8.3, 8.65, 9.1, 10.6, 11.3]
TIR = np.array(TIR)
astfusedtirwave = np.array(astfusedtirwave)
astfusedtirrefchl = np.array(astfusedtirrefchl)

TIRrefchl = spectres(TIR, astfusedtirwave, astfusedtirrefchl)
```

```python
# Fill this whole kernal out for each new mineral to give VNIR - SWIR and TIR reflectance

# Read resampled ASTER and Sentinel-2 reflectance and NIC NICOLET 1.12 - 216µm raw reflectance values

AST_REF = open('S07ASTER_Calcite_WS272_ASDNGa_AREF.txt')

astfusedrefcal = AST_REF.readlines()[4:10]
for i in range (0, len(astfusedrefcal)):
    astfusedrefcal[i] = float(astfusedrefcal[i])
    
SENT_REF = open('S07SNTL2_Calcite_WS272_ASDNGa_AREF.txt')

sentfusedrefcal = SENT_REF.readlines()[2:10]
for i in range(0, len(sentfusedrefcal)):
    sentfusedrefcal[i] = float(sentfusedrefcal[i])
    
AST_TIR_REF = open('splib07a_Calcite_WS272_NIC4aaa_RREF.txt')

astfusedtirrefcal = AST_TIR_REF.readlines()[922:]
for i in range (0, len(astfusedtirrefcal)):
    astfusedtirrefcal[i] = float(astfusedtirrefcal[i])

#-----------------------------------------------------------------------------------------     

# Order the VNIR - SWIR bands by increasing corresponding wavelength

fusedwave = astfusedwave + sentfusedwave
fusedrefcal = astfusedrefcal + sentfusedrefcal

zipped_lists = zip(fusedwave, fusedrefcal)
sorted_pairs = sorted(zipped_lists)

tuples = zip(*sorted_pairs)
fusedwavelength, fusedrefcal = [ list(tuple) for tuple in  tuples]

#----------------------------------------------------------------------------------------- 

# Resample the NIC NICOLET 1.12 - 216µm reflectance data to ASTER TIR bands

TIR = [8.3, 8.65, 9.1, 10.6, 11.3]
TIR = np.array(TIR)
astfusedtirwave = np.array(astfusedtirwave)
astfusedtirrefcal = np.array(astfusedtirrefcal)

TIRrefcal = spectres(TIR, astfusedtirwave, astfusedtirrefcal)
```

```python
# This kernal is to calculate the fused bare earth composite reflectance for pyrite from BEC (VNIR to SWIR) and NIC (TIR) as there were none available (no ASD)

fusedwaves = astfusedwave + sentfusedwave
fusedwaves = np.array(fusedwaves)

# Use BEC reflectance dataset to calculate reflectacne in the fused VNIR to SWIR bands
BEC_PYR_VNIRtoSWIR = open('splib07a_Pyrite_HS35.3_BECKb_AREF.txt')
becpyrite = BEC_PYR_VNIRtoSWIR.readlines()[1:]
for i in range(0, len(becpyrite)):
    becpyrite[i] = float(becpyrite[i])
becpyrite = np.array(becpyrite)

# Calculate the resampled reflectance
resampled_pyrite_VNIRtoSWIR_flux = spectres(fusedwaves, becwaves, becpyrite)
```

```python
# Define the array to plot the wavelengths of the fused bare earth composite bands for visualisation purposes

FUSED_arr = []
for i in range (0,len(fusedwavelength)):
    FUSED_arr.append([fusedwavelength[i], fusedwavelength[i]])

for i in range (0, len(TIR)):
    FUSED_arr.append([TIR[i], TIR[i]])

Vert = [0,1]
```

```python
# Create figure and set size
pl.figure(figsize=(15,8))

# Plot lines (add plot for each new mineral)
pl.plot(fusedwavelength, fusedrefchl, 'lightgreen', label = 'Chlorite')
pl.plot(TIR, TIRrefchl, 'lightgreen')

pl.plot(fusedwavelength, fusedrefepi, 'b', label = 'Epidote')
pl.plot(TIR, TIRrefepi, 'b')

pl.plot(fusedwavelength, fusedrefact, 'g', label = 'Actinolite')
pl.plot(TIR, TIRrefact, 'g')

pl.plot(fusedwavelength, fusedrefcal, 'gray', label = 'Calcite')
pl.plot(TIR, TIRrefcal, 'gray')

pl.plot(fusedwavelength, resampled_pyrite_VNIRtoSWIR_flux, 'darkkhaki', label = 'Pyrite')

# Set limits for the whole plot
pl.xlim(0.2,2.5)
pl.ylim(0,1)
pl.xlabel('Wavelength (µm)', fontsize = 18)
pl.ylabel('Reflectance', fontsize = 18)
pl.title('Propylitic', fontsize = 26)
pl.legend()

for i in range(len(FUSED_arr)):
    pl.plot(FUSED_arr[i], Vert, 'k--', label = 'ion abs', alpha = 0.3)

pl.savefig('Propylitic')
    
pl.show()
```

```python
# Export .csv file of resampled reflectance data

for i in range (0,len(TIRrefchl)):
    fusedrefchl.append(TIRrefchl[i])

import numpy
a = numpy.zeros(shape=(len(fusedrefchl), 1))

for i in range (0,len(fusedrefchl)):
    a[i] = fusedrefchl[i]

import csv 
file = open('Chlorite Reflectance.csv', 'w+', newline ='')  
with file:     
    write = csv.writer(file) 
    write.writerows(a) 
```
